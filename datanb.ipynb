```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
print(data.columns)
```

```python
X = pd.DataFrame(data,columns=['GENDER', 'AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY',
       'PEER_PRESSURE', 'CHRONIC DISEASE', 'FATIGUE ', 'ALLERGY ', 'WHEEZING',
       'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH'])

y = pd.DataFrame(data, columns=['LUNG_CANCER'])


train_x, test_x, train_Y, test_Y = train_test_split(X, y, test_size=.3, shuffle=True,random_state=119)
```

```python
#print(X.value_counts)
from sklearn.metrics import accuracy_score, classification_report

print(y.value_counts)

model = RandomForestClassifier(n_estimators=100, random_state=119)
#train_x, train_y, test_X, test_Y 
model.fit(train_x, train_Y)


predictions = model.predict(test_x)

y_pred1 = accuracy_score(predictions, test_Y)
y_pred2 = classification_report(predictions, test_Y)

print(y_pred1)
print(y_pred2)
```
